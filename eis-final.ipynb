### Prepare RNA data with Image data

```python
import os
import pydicom
import numpy as np
import cv2
import pandas as pd

base_path = r"C:\Users\hudda\OneDrive\Desktop\EIS"

dataset_folders = [
    "NSCLC Radiogenomics-20260326T124713Z-3-001",
    "NSCLC Radiogenomics-20260326T124713Z-3-002",
    "NSCLC Radiogenomics-20260326T124713Z-3-003",
    "NSCLC Radiogenomics-20260326T124713Z-3-004"
]

# -------- LOAD RNA --------
rna_path = os.path.join(
    base_path,
    dataset_folders[0],
    "NSCLC Radiogenomics",
    "merged_rna_seq.csv"
)

df_rna = pd.read_csv(rna_path)

# -------- FIND COMMON IDS --------
image_ids = []

for dataset in dataset_folders:
    root_path = os.path.join(base_path, dataset, "NSCLC Radiogenomics")
    
    for folder in os.listdir(root_path):
        if os.path.isdir(os.path.join(root_path, folder)):
            image_ids.append(folder)

common_ids = set(image_ids).intersection(set(df_rna["Case ID"]))

# -------- BUILD IMAGE DATA --------
all_patients = []
processed_ids = set()
valid_ids = []

for dataset in dataset_folders:
    root_path = os.path.join(base_path, dataset, "NSCLC Radiogenomics")
    
    for folder in os.listdir(root_path):
        folder_path = os.path.join(root_path, folder)
        
        if not os.path.isdir(folder_path):
            continue
        
        if folder not in common_ids or folder in processed_ids:
            continue
        
        processed_ids.add(folder)
        
        slices = []
        
        for f in sorted(os.listdir(folder_path)):
            if not f.endswith(".dcm"):
                continue
            
            try:
                ds = pydicom.dcmread(os.path.join(folder_path, f))
                img = ds.pixel_array
                
                img = (img - np.min(img)) / (np.max(img) - np.min(img) + 1e-8)
                
                if len(img.shape) == 3:
                    img = img[:, :, 0]
                
                img = cv2.resize(img, (224, 224))
                slices.append(img)
                
            except:
                continue
        
        if len(slices) < 60:
            continue
        
        slices = np.stack(slices)
        
        mid = len(slices) // 2
        selected = slices[mid-25:mid+25]
        selected = selected[..., np.newaxis]
        
        all_patients.append(selected)
        valid_ids.append(folder)

# -------- FINAL IMAGE DATA --------
X_images = np.array(all_patients)

print("X_images shape:", X_images.shape)
print("Valid patients:", len(valid_ids))

# -------- RNA ALIGNMENT --------
df_rna_final = df_rna[df_rna["Case ID"].isin(valid_ids)]
df_rna_final = df_rna_final.set_index("Case ID").loc[valid_ids]

X_rna = df_rna_final.values

print("X_rna shape:", X_rna.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split

# -------- CONVERT TO RGB --------
X_img_rgb = np.repeat(X_img, 3, axis=-1)

# -------- TRAIN TEST SPLIT --------
X_train_img, X_test_img, y_train_img, y_test_img = train_test_split(
    X_img_rgb, y_img, test_size=0.2, random_state=42
)

# -------- LOAD PRETRAINED MODEL --------
base_model = MobileNetV3Small(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

# freeze base (important for small dataset)
base_model.trainable = False

# -------- ADD CLASSIFIER --------
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(64, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model_mobilenet = Model(inputs=base_model.input, outputs=output)

# -------- COMPILE --------
model_mobilenet.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# -------- TRAIN --------
history = model_mobilenet.fit(
    X_train_img, y_train_img,
    validation_data=(X_test_img, y_test_img),
    epochs=10,
    batch_size=8,
    verbose=1
)

# -------- EVALUATE --------
loss, acc = model_mobilenet.evaluate(X_test_img, y_test_img)

print("MobileNetV3 Accuracy:", acc)
```

### MobileNetV3 + RNA Multimodal

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.layers import (
    Input, Dense, GlobalAveragePooling2D,
    Concatenate, TimeDistributed, Lambda
)
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# -------- PREPARE IMAGE DATA --------
# grayscale → RGB
X_img_rgb = np.repeat(X_images, 3, axis=-1)   # (39, 50, 224, 224, 3)

# -------- TRAIN TEST SPLIT --------
X_img_train, X_img_test, X_rna_train, X_rna_test, y_train, y_test = train_test_split(
    X_img_rgb, X_rna, y_final,
    test_size=0.2,
    random_state=42
)

# -------- SCALE RNA --------
scaler = StandardScaler()
X_rna_train = scaler.fit_transform(X_rna_train)
X_rna_test = scaler.transform(X_rna_test)

# -------- IMAGE MODEL (MobileNetV3) --------
base_model = MobileNetV3Small(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

# freeze base model (important for small dataset)
base_model.trainable = False

# -------- IMAGE INPUT --------
image_input = Input(shape=(50, 224, 224, 3))

# apply CNN per slice
x = TimeDistributed(base_model)(image_input)
x = TimeDistributed(GlobalAveragePooling2D())(x)

# -------- FIXED PART (aggregation) --------
x = Lambda(lambda x: tf.reduce_mean(x, axis=1))(x)

# dense features
image_features = Dense(64, activation='relu')(x)

# -------- RNA MODEL --------
rna_input = Input(shape=(11173,))

r = Dense(64, activation='relu')(rna_input)
r = Dense(32, activation='relu')(r)

# -------- FUSION --------
combined = Concatenate()([image_features, r])

z = Dense(32, activation='relu')(combined)
z = Dense(1, activation='sigmoid')(z)

# -------- FINAL MODEL --------
model_mobilenet_multi = Model(
    inputs=[image_input, rna_input],
    outputs=z
)

model_mobilenet_multi.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_mobilenet_multi.summary()

# -------- TRAIN --------
history = model_mobilenet_multi.fit(
    [X_img_train, X_rna_train],
    y_train,
    validation_data=([X_img_test, X_rna_test], y_test),
    epochs=10,
    batch_size=4,
    verbose=1
)

# -------- EVALUATE --------
loss, acc = model_mobilenet_multi.evaluate(
    [X_img_test, X_rna_test], y_test
)

print("Multimodal MobileNet Accuracy:", acc)
```
